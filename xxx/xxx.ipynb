### 개념 정리
양액: 식물 재배에 필요한 질소, 인산, 칼륨 등 필수 양분을 물에 녹여 만든 액상 비료 또는 영양액
배지: 흙 대신 식물 뿌리를 잡아주는 재료
공급 EC: 식물에게 주기 직전 양액의 농도
배지 EC: 배지(코코피트 등) 안에 실제로 존재하는 염류 농도
        -> 시간이 지나면 배지 EC가 올라갈 수 있음
배액 EC: 식물 뿌리를 거쳐 나온 물의 농도
        -> 식물이 영양을 얼마나 먹었는지 알 수 있음
공급 pH: 식물에게 주는 양액의 산도
배액 pH: 뿌리를 지나 나온 물의 산도
토양 EC: 토양 속에 녹아 있는 염류(영양분)의 농도
온도1 / 온도2: 서로 다른 위치에서 측정한 실내 공기 온도 센서값
난방관1온도 / 난방관2온도: 난방 배관(파이프) 표면 온도
                -> 공기 온도가 아니라 난방 시스템 자체의 온도
난방설정온도: 목표온도
이슬점온도: 공기 중 수분이 응결되기 시작하는 온도
                -> 공기중 수증기량에 의존적인 데이터
예비온도: 보조 목표온도 
                -> 주 센서가 고장났을 때 쓰는 온도
3 way: 3방 밸브, 물이나 온수의 흐름 방향을 조절하는 밸브
                ->밸브 개도율(%), 흐름을 정밀 제어하기 위해
순환펌프: 온수를 배관 안에서 계속 돌려주는 펌프
훈증기: 연무(안개) 형태로 약제 살포(병해충 방제, 소독)
유동팬: 내부 공기 순환용 팬
배기팬: 외부로 공기를 빼내는 팬
보광: 보조 조명, 일조 부족 시 빛 보충

### 고민사항
이 농가는 배지 재배인가, 토양 재배인가?
-> 식물 생육에 직접 영향을 주는 것은 "뿌리가 위치한 환경"이기 때문

A. 배지 재배라면

뿌리 → 코코피트/락울 등 배지에 위치
배지EC, 배지수분이 핵심 변수
토양EC는 보조 모니터링 지표 가능성
-> 이 경우 NT가 핵심 데이터

B. 토양 재배라면

뿌리 → 흙에 직접 위치
토양EC, 토양PH가 핵심 변수
배지EC라는 명칭이 실제로는 토양센서일 가능성도 있음
-> 이 경우 EL이 핵심 데이터

### 분석가능 방법
Energy와 Environment의 관계 분석
→ 에너지 사용이 온실 내부 환경 조건에 어떠한 영향을 미치는지 분석한다.
Environment와 Growth의 관계 분석
→ 내부 환경 요인이 작물의 생육 지표에 미치는 영향을 분석한다.
Growth와 Production의 관계 분석
→ 생육 지표가 실제 수확량 및 상품 등급에 어떻게 연결되는지 분석한다.

위의 분석을 통해 설명가능한 조합
1. 특정 환경 조건(온도, 습도, CO₂ 등)을 유지하기 위해 어느 정도의 에너지가 필요한가?
→ 에너지 대비 환경 제어 효율 분석
2. 식물의 성장을 촉진하는 핵심 환경 요인은 무엇인가?
→ 생육에 유의미한 환경 변수 도출
3. 상품의 품질 및 생산량은 어떤 생육 지표와 밀접한 관련이 있는가?
→ 생육 지표와 수확량/등급 간의 구조적 관계 분석
4. 특정 생육 수준을 달성하기 위해 필요한 에너지 투입 수준은 얼마인가?
→ 에너지 → 환경 → 생육 → 생산의 연쇄적 영향 분석
5. 동일한 생산량을 달성하는 데 있어 에너지 사용 효율이 높은 운영 패턴은 무엇인가?
→ Yield 대비 Energy 사용량 분석을 통한 효율적 운영 전략 도출
6. 환경 변동성(예: 온도 변동 폭)이 생육 및 생산 안정성에 미치는 영향은 무엇인가?
→ 평균값이 아닌 ‘환경 안정성’의 중요성 검증 -> std
7. 에너지 투입에는 수확량 증가에 대한 한계가 존재하는가?
→ 과도한 에너지 사용의 비효율성 검증
